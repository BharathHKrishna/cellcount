# Weekend 1 — Cellpose baseline

Run pretrained Cellpose on a sample image, get an instance mask, count cells, draw outlines. Prove the v1 pipeline end to end before wrapping it in Gradio.

```python
import sys
sys.path.append("../app")

from segment.cellpose_runner import run_cellpose
from segment.postprocess import measure_instances
from segment.overlay import draw_outlines
```

```python
import matplotlib.pyplot as plt
from skimage import data

image = data.human_mitosis()  # real fluorescence nuclei micrograph, from scikit-image
plt.figure(figsize=(5, 5))
plt.imshow(image, cmap="gray")
plt.title(f"input image {image.shape}")
plt.axis("off")
plt.show()
```

```python
masks = run_cellpose(image)
print(f"cellpose found {masks.max()} instances")
```

```python
measurements = measure_instances(masks)
overlay = draw_outlines(image, masks)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image, cmap="gray")
axes[0].set_title("original")
axes[0].axis("off")
axes[1].imshow(overlay)
axes[1].set_title(f"{len(measurements)} cells detected")
axes[1].axis("off")
plt.tight_layout()
plt.show()
```

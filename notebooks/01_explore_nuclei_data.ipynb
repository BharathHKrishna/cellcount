# Explore the 2018 Data Science Bowl / BBBC038 nuclei dataset

Look at a handful of images + their instance masks before writing the training dataset loader.

```python
import sys
sys.path.append("../train")

from dataset import DSBNucleiDataset
```

```python
ds = DSBNucleiDataset("../data/BBBC038/stage1_train", image_size=256)
print(f"{len(ds)} training images")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(14, 7))
for i in range(4):
    image, mask = ds[i * 50]
    axes[0, i].imshow(image.permute(1, 2, 0))
    axes[0, i].set_title(f"image {i * 50}")
    axes[0, i].axis("off")
    axes[1, i].imshow(mask[0], cmap="gray")
    axes[1, i].set_title("binary mask")
    axes[1, i].axis("off")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

foreground_fractions = [ds[i][1].mean().item() for i in range(0, len(ds), 20)]
print(f"mean foreground fraction: {np.mean(foreground_fractions):.3f}")
plt.hist(foreground_fractions, bins=20)
plt.xlabel("foreground fraction per image")
plt.ylabel("count")
plt.title("Class balance across the dataset (subsampled)")
plt.show()
```
